```python
# Monjaraz Briseño Luis Fernando
# Extra ejercicio A sin numpy

import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo de datos
datos_completos = pd.read_csv('/car_purchasing.csv', encoding='latin-1', header=0)
datos = datos_completos[['annual Salary', 'car purchase amount']]

# Variables X (independiente) y Y (dependiente)
sueldo = datos['annual Salary']
compra_auto = datos['car purchase amount']

# Calcular las medias de X e Y
sueldo_medio = sum(sueldo) / len(sueldo)
compra_auto_media = sum(compra_auto) / len(compra_auto)

# Calcular la pendiente (m) manualmente
numerador = sum((sueldo[i] - sueldo_medio) * (compra_auto[i] - compra_auto_media) for i in range(len(sueldo)))
denominador = sum((sueldo[i] - sueldo_medio) ** 2 for i in range(len(sueldo)))
pendiente = numerador / denominador

# Calcular la ordenada al origen (b)
intercepto = compra_auto_media - pendiente * sueldo_medio

# Realizar predicciones
prediccion_compra = [pendiente * x + intercepto for x in sueldo]

# Graficar los puntos de datos y la línea de regresión
plt.scatter(sueldo, compra_auto, color='gold', label='Datos')
plt.plot(sueldo, prediccion_compra, color='red', label='Regresión Lineal')

# Agregar la media como un punto
plt.scatter(sueldo_medio, compra_auto_media, color='blue', marker='s', label='Media', zorder=5)

# Etiquetas de los ejes y leyenda
plt.xlabel('Sueldo Anual/Variable independiente')
plt.ylabel('Monto de Compra de Auto/Variable dependiente')
plt.legend()
plt.show()

# Imprimir la pendiente (m) y la ordenada al origen (b)
print(f'Pendiente (m): {pendiente}')
print(f'Ordenada al origen (b): {intercepto}')
```

```python
# Monjaraz Briseño Luis Fernando

import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo de datos
datos = pd.read_csv('/car_purchasing.csv', usecols=[5, 8], encoding='latin-1')

# División de los datos en 60% para entrenamiento y 40% para prueba
entrenamiento = datos.sample(frac=0.6, random_state=0)
prueba = datos.drop(entrenamiento.index)

# Variables X (independiente) y Y (dependiente) para el conjunto de entrenamiento
sueldo_entrenamiento = entrenamiento['annual Salary']
compra_entrenamiento = entrenamiento['car purchase amount']

# Calcular la media de X e Y
sueldo_entrenamiento_medio = sueldo_entrenamiento.mean()
compra_entrenamiento_media = compra_entrenamiento.mean()

# Calcular la covarianza manualmente
covarianza = sum((sueldo_entrenamiento.iloc[i] - sueldo_entrenamiento_medio) *
                 (compra_entrenamiento.iloc[i] - compra_entrenamiento_media)
                 for i in range(len(sueldo_entrenamiento))) / len(sueldo_entrenamiento)

# Calcular la varianza de X manualmente
varianza = sum((sueldo_entrenamiento.iloc[i] - sueldo_entrenamiento_medio) ** 2
               for i in range(len(sueldo_entrenamiento))) / len(sueldo_entrenamiento)

# Pendiente (m) con la fórmula de: m = cov(x, y) / var(x)
pendiente = covarianza / varianza

# Intercepto (b) con la fórmula de: b = y - m * x
intercepto = compra_entrenamiento_media - pendiente * sueldo_entrenamiento_medio

# Predicciones en el conjunto de prueba
sueldo_prueba = prueba['annual Salary']
compra_prueba = prueba['car purchase amount']
predicciones = [pendiente * x + intercepto for x in sueldo_prueba]

# Graficar los datos de entrenamiento y la línea de regresión
plt.scatter(sueldo_entrenamiento, compra_entrenamiento, color='blue', label='Datos de Entrenamiento')
plt.plot(sueldo_entrenamiento, [pendiente * x + intercepto for x in sueldo_entrenamiento], color='red', label='Regresión Lineal')
plt.xlabel('Sueldo Anual/Variable independiente')
plt.ylabel('Monto de Compra de Auto/Variable dependiente')
plt.legend()
plt.show()

# Graficar los datos de prueba y la línea de regresión
plt.scatter(sueldo_prueba, compra_prueba, color='gold', label='Datos de Prueba')
plt.plot(sueldo_entrenamiento, [pendiente * x + intercepto for x in sueldo_entrenamiento], color='red', label='Regresión Lineal')
plt.xlabel('Sueldo Anual/Variable independiente')
plt.ylabel('Monto de Compra de Auto/Variable dependiente')
plt.legend()
plt.show()

# Imprimir la pendiente (m) y el intercepto (b)
print(f'Pendiente (m): {pendiente}')
print(f'Intercepto (b): {intercepto}')
```
